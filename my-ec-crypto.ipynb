### Secp256k1 (ethereum)
$y^2 = x^3 + 7$

```python
import math
import matplotlib.pyplot as plt
!pip3 install pycryptodome

#  y = sqrt(x^3 + 7)
def secp(x):
    w = abs(x**3 + 7)    
    return math.sqrt(w)


class SecpPlot:

    def __init__(self):
        xs = list(map(lambda x: x/100.0,range(-190,500,1)))
        ys = list(map(secp,xs))

        self.fig, self.ax = plt.subplots(figsize=(4,7)) 

        self.ax.plot(xs,ys)

        self.ax.plot(xs, list(map(lambda x: -x,ys)))
    
    def show(self):
        self.ax.spines['left'].set_position('zero')
        self.ax.spines['bottom'].set_position('zero')
        self.fig.show(False)
    
    def point(self, point, letter,color='red'):
        self.ax.plot(point.x,point.y, markersize=5,color=color,marker='o')
        self.ax.text(point.x, point.y, letter, fontsize=12, verticalalignment='bottom', horizontalalignment='right')

        self.fig.show(False)

plot = SecpPlot()
plot.show()
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __neg__(self):
        return Point(self.x,-self.y)

    # this is defined geometrically
    def __add__(self,b):
        if self == self.infinity():
            return Point(b.x,b.y)
        if b == self.infinity():
            return Point(self.x,self.y)
        if self == -b:
            return self.infinity()
        if self == b:
            return self.double()
        else:
            slope = (b.y - self.y)/(b.x - self.x)
            c_x = slope**2 - self.x - b.x
            c_y = slope * (self.x - c_x) - self.y
            return Point(c_x, c_y)

    def double(self):
        slope = (self.x **2 * 3)/ (self.y *2)
        c_x = slope**2 - self.x - self.x
        c_y = slope * (self.x - c_x) - self.y
        return Point(c_x, c_y)

    def __eq__(self,other):
        return self.x == other.x and self.y == other.y

    def __repr__(self):
        return f"Point({self.x}, {self.y})"

    def infinity(self):
        return Point(float('inf'),float('inf'))
```

```python
# different points addition
A = Point(2, secp(2))
B = Point(3, secp(3))
C = A+B
plot = SecpPlot()
plot.point(A, 'A')
plot.point(B, 'B')
print(f"{C=}")
plot.point(C, 'C')
plot.show()
```

```python
# field element of order 41
# 5 / 10 => 5 * 10^-1 => 10^-1 10  
# 10 * 10^-1 mod N = 1 
def extended_euclid(prev_r, r):
    assert prev_r >= r
    prev_s, s = 1, 0
    prev_t, t = 0, 1
    while (r != 0):
        quotient = prev_r // r
        prev_r, r = r, prev_r - quotient * r
        prev_s, s = s, prev_s - quotient * s
        prev_t, t = t, prev_t - quotient * t
    return prev_t

def create_fe_class(order):
    class FE:
        def __init__(self, value):    
            if isinstance(value,FE):
                self.value = value.value
            elif value == 'inf':
                self.value = 0
            else:
                self.value = value % order
          
        def __add__(self, other):
            return FE(self.value + other.value)
          
        def __sub__(self, other):
            return FE(self.value - other.value)
    
        def __mul__(self, other):
            return FE(self.value * FE(other).value)
    
        def __pow__(self, other):
            assert isinstance(other,int)
            return FE(self.value ** other)
    
        def __truediv__(self, other):
            inv = FE(extended_euclid(order,other.value))
            return self * inv
                
        def __neg__(self):
            return FE(-self.value)
        
        def __eq__(self, other):
            return self.value == other.value
    
        def __repr__(self):
            return f"FE({self.value})"
        
    return FE
FE = create_fe_class(41)

assert FE(5) + FE(6) == FE(11)
assert FE(30) + FE(12) == FE(1)
assert FE(30) + FE(-12) == FE(18)
assert FE(12) - FE(30) == FE(41-18)
assert FE(10) * FE(10) == FE(18)
assert FE(10) * FE(10) == FE(10) ** 2
assert FE(10) * FE(10) * FE(10) == FE(10) ** 3
assert FE(10) / FE(5) == FE(2)
assert FE(2) * FE(5) == FE(10)
assert FE(22) * FE(32) == FE(130)
```

```python
def create_fe_point_class(inf_element):
    class FEPoint(Point):
        def infinity(self):
            return FEPoint(inf_element,inf_element)
    
        def scalar_mul(self,n):
            acc = self
            result = self.infinity()
            for bit in bin(n)[2:][::-1]:
                if bit == '1':
                    result = result + acc
                acc = acc.double()
            return result
    
        def __neg__(self):
            return FEPoint(self.x,-self.y)
    
        # this is defined geometrically
        def __add__(self,b):
            if self == self.infinity():
                return FEPoint(b.x,b.y)
            if b == self.infinity():
                return FEPoint(self.x,self.y)
            if self == -b:
                return self.infinity()
            if self == b:
                return self.double()
            else:
                slope = (b.y - self.y)/(b.x - self.x)
                c_x = slope**2 - self.x - b.x
                c_y = slope * (self.x - c_x) - self.y
                return FEPoint(c_x, c_y)
    
        def double(self):
            if self == self.infinity():
                return self
            if self.y.value == 0:
                return self.infinity()
            slope = ((self.x **2) * 3)/ (self.y *2)
            c_x = slope**2 - self.x - self.x
            c_y = slope * (self.x - c_x) - self.y
            return FEPoint(c_x, c_y)
    
        def __repr__(self):
            return f"FEPoint({self.x}, {self.y})"
    return FEPoint
FEPoint = create_fe_point_class(FE(0))        
INF = FEPoint( FE(0), FE(0))

assert INF + FEPoint(FE(1), FE(2)) == FEPoint(FE(1), FE(2))
assert FEPoint(FE(1), FE(2)) + INF == FEPoint(FE(1), FE(2))
assert FEPoint(FE(1), FE(2)) + FEPoint(FE(1), FE(-2)) == INF
```

```python
import random
from Crypto.Hash import keccak

secp256k1 = {
    'order': 0xFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFEBAAEDCE6AF48A03BBFD25E8CD0364141,
    'prime_field_p': 0xfffffffffffffffffffffffffffffffffffffffffffffffffffffffefffffc2f,
    'generator_x': 0x79be667ef9dcbbac55a06295ce870b07029bfcdb2dce28d959f2815b16f81798, 
    'generator_y': 0x483ada7726a3c4655da4fbfc0e1108a8fd17b448a68554199c47d08ffb10d4b8
}
Secp256k1Element = create_fe_class(secp256k1['prime_field_p'])
Secp256k1Point = create_fe_point_class(Secp256k1Element(0))
Secp256k1G = Secp256k1Point(Secp256k1Element(secp256k1['generator_x']),Secp256k1Element(secp256k1['generator_y']))

def int_to_hex_bytes(x):
    hex_str = hex(x)[2:]
    if len(hex_str) % 2 ==0:
        return bytes.fromhex(hex_str)
    return bytes.fromhex('0'+hex_str)

class Wallet:
    def __init__(self, private_key=None):
         if private_key == None:
             private_key = random.randint(1,secp256k1['order']) 
         self.private_key = private_key
         self.public_key_point = Secp256k1G.scalar_mul(self.private_key)

    def __repr__(self):
        k = keccak.new(digest_bits=256)
        k.update(int_to_hex_bytes(self.public_key_point.x.value))
        k.update(int_to_hex_bytes(self.public_key_point.y.value))
        address = '0x' + k.hexdigest()[-40:]
        return f"private_key={hex(self.private_key)} address={address}"
```

```python
wallet = Wallet(0x765f147f34336f1e2a8d9d4abb86fc50f4c68632924d01de89777d92449575e1)

def eth_like_sign(wallet, msg):
    e = keccak.new(digest_bits=256).update(msg).hexdigest()
    ln = len(bin(secp256k1['order'])) - 2
    z = int(e[:ln],16) # in practice always: len(z) == len(e)
    k = random.randint(1,secp256k1['order'])
    p_k = Secp256k1G.scalar_mul(k)
    x,y = (p_k.x,p_k.y)
    r = x.value % secp256k1['order']
    if r == 0:
        raise "Very rare case r == 0"
    s = ( (z + r * wallet.private_key) * pow(k,-1,secp256k1['order']) ) % secp256k1['order']
    if r == 0:
        raise "Very rare case s == 0"
    v = y.value % 2

    if s * 2 >= secp256k1['order']:
            s = secp256k1['order'] - s
            v ^= 1
        
    return (r,s,v + 27)

def eth_like_verify(msg, sig, pk):
    assert pk != Secp256k1G.infinity() # not inf
    assert pk.y**2 == pk.x**3 + Secp256k1Element(7) # is on curve
    assert pk.scalar_mul(secp256k1['order'])
    
    r,s,v = sig
    assert r > 1 and r < secp256k1['order']
    assert s > 1 and s < secp256k1['order']
    
    e = keccak.new(digest_bits=256).update(msg).hexdigest()
    ln = len(bin(secp256k1['order'])) - 2
    z = int(e[:ln],16) # in practice always: len(z) == len(e)
    
    u_1 = z * pow(s,-1,secp256k1['order']) % secp256k1['order']
    u_2 = r * pow(s,-1,secp256k1['order']) % secp256k1['order']

    p = Secp256k1G.scalar_mul(u_1) + pk.scalar_mul(u_2)
    if p == Secp256k1G.infinity():
        return False

    return r == p.x.value % secp256k1['order']
    
    

sig = eth_like_sign(wallet, b'ala ma kota')
assert eth_like_verify(b'ala ma kota', sig, wallet.public_key_point) == True
assert eth_like_verify(b'ala ma kot2', sig, wallet.public_key_point) == False
assert eth_like_verify(b'ala ma kota', sig, Wallet().public_key_point) == False
```
